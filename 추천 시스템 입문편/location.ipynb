### Pandas Lambda와 Apply를 이용해 간단하게 풀어보기

```python
import pandas as pd
```

```python
movies = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/추천 시스템 입문편/data/ml-latest-small/movies.csv', index_col='movieId')
movies.sample()
```

```python
# apply는 for문의 기능을 대신해줌
# lambda는 간단한 함수를 만들지 않아도 되도록 함
# apply(함수) : 함수를 lambda 함수로도 써 줄 수 있음
movies['year'] = movies['title'].apply(lambda x: x[-5:-1])
# 개봉연도 정제 오류 발생
movies['year'].unique()
```

### Location을 이용해 정제 오류 고치기

```python
movies[movies['year'] == '012)']
```

```python
# movies.loc[index, column]
# 공백이 한 칸 들어가 있는 것을 알 수 있음
movies.loc[102084, 'title']
```

```python
# 문자열의 앞 뒤 공백을 없애도록 함
movies.loc[102084, 'title'].strip()
```

```python
# 전체 데이터의 앞 뒤 공백을 없애줌
movies['title'] = movies['title'].apply(lambda x: x.strip())
movies
```

```python
movies['year'] = movies['title'].apply(lambda x: x[-5:-1])
# 개봉연도 정제 오류가 조금 수정됨
movies['year'].unique()
```

```python
# 이상한 값들을 모음
errors = ['lon ', 'r On', ' Roa', 'atso', 'imal', 'erso', 'ligh', 'he O', 'osmo', ' Bab', 'ron ', 'irro']
```

```python
# 에러가 있는 데이터들만 뽑아봄
# 개봉 연도가 없는 것을 볼 수 있음
movies[movies['year'].isin(errors)]
```

```python
# movies.loc[필터링정보, column]
# year 컬럼 정보만 뽑아오도록 함
# 모두 2050년에 개봉했다고 변경
movies.loc[movies['year'].isin(errors), 'year'] = '2050'
```

```python
# 개봉연도 정제 오류가 모두 수정됨
movies['year'].unique()
```
